### Taxa de Creixement

Ara, gràficarem la funció de taxa de creixement (p'/p). Estudiarem models continus de la forma 


$\left\lbrace\begin{array}{c}p'=pf(p) \\ p(t_0)=p_0 \end{array}\right.$

En la següent gràfica utilitzarem una aproximació 

$p'(t)=\frac{p(t+h)-p(t)}{h}$

```python
dY=[(Yt[i+1]-Yt[i])/(Xt[i+1]-Xt[i]) for i in range(0,len(Xt)-1)];
dYY=[dY[i]/Yt[i] for i in range(0,len(dY))];
taula0=points([[Yt[i],dYY[i]] for i in range(0,len(dY))]);
show(taula0);
```

Un cop tenim la funció de taxa de creixement, podem pasar a estudiar els models continus i comparar-los amb la població en cada any.

# Model de Verhulst I

En aquest cas, trobem un model amb taxa de creixement lineal 

$f(p)=ap+b$

Busquem la recta de regressió que més s'ajusta al conjunt de punts (p, p'/p) per tal de seguir el model

El conjunt de dades serà (p, p'/p)

```python
YtdYY=[[Yt[i],dYY[i]] for i in range(0,len(dY))]; YtdYY
```

Trobem la gràfica

```python
var("a b p2")
```

```python
model(p2)=a*p2+b
```

```python
recta0=find_fit(YtdYY,model)
```

```python
recta0
```

```python
points(YtdYY)+plot(model(a=recta0[0].rhs(),b=recta0[1].rhs()),(p2,38000, 165000),color="red")
```

Trobem p, sabent que $p'=p(ap+b)$

```python
var('y')
```

```python
integral(1/(y*(a*y+b)), y)
```

```python
a= recta0[0].rhs()
b=recta0[1].rhs()
b
```

```python
sols=solve([log(y/(a*y+b))-log(Yt[0]/(a*Yt[0]+b))==b*(t-Xt[0])],y);sols
```

```python
PlotVerhust=plot(sols[0](t).rhs(),(t,1950,2020), color='red', legend_label="$Model Verhust$");PlotVerhust+taula
```

Es veu a la vista que aquest model millora considerablement el model de Maldust
