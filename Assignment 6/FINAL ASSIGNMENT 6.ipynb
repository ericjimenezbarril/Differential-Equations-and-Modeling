### Interpretació del resultat del model predador-presa

Dibuixem el camp vectorial a prop dels punts crítics trobats amb les dues components positives, ja que estem estudiant el model predador-presa on x,y representen poblacions.

```python
P1=plot_vector_field([eq1,eq2],[x,0,3],[y,0,3])
plot(P1)
```

En aquest cas $x(t) = \textit{nombre de preses} \geq 0$ i $y(t)=\textit{nombre de predadors} \geq 0$. 

És així pel següent raonament: si suposem que tenim molts prededors i poques preses (y>>0 i x $\approx 0$) aleshores el número de prededors disminueix perquè no tenen aliment. Quan y $\approx 0$, les preses no són amenaçades pels prededors i per tant poden sobreviure. Aleshores el número de preses augmenta (x>>0). Un cop tenim moltes preses, el número de predadors augmenta ja que tenen aliment per sobreviure. D'aquesta manera tanquem el cicle, que correspon al retrat de fase mostrat. 
És per això que $x(t) = \textit{nombre de preses} \geq 0$ i $y(t)=\textit{nombre de predadors} \geq 0$.

Ara, com coneixem quin tipus de punt es cadascun dels punts singulars i sabem que només em de estudiar el primer quadrant, estudiem $\textit{les separatrius}$ del primer quadrant
