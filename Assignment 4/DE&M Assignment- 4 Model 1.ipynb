```python
# Model 1 
# Mostrar que la corba que té la propietat de que totes
# les seves normals passen per un punt fix, és una circumferència.
```

```python
f(x)=sqrt(1-x);plot0=plot(f(x),0,1);plot0
```

```python
x0=1/2;y0=f(x=x0);der=diff(sqrt(1-x),x);yp0=der.substitute(x=x0)
```

```python
plot1=plot(yp0*(x-x0)+y0,0,1,color="red");plot1
```

```python
plot0+plot1
```

```python
plot2=plot((-1/yp0)*(x-x0)+y0,0,1,color="green");plot0+plot1+plot2
```

```python
plot0=plot(f(x),0,1,aspect_ratio=1);
plot1=plot(yp0*(x-x0)+y0,0,1,color="red",aspect_ratio=1);
plot2=plot((-1/yp0)*(x-x0)+y0,0,1,color="green",aspect_ratio=1);
plot0+plot1+plot2
```

```python
x1=3/4;
y1=f(x=x1)
yp1=der.substitute(x=x1)
plot3=plot(yp1*(x-x1)+y1,0,1,color="red",aspect_ratio=1);
plot4=plot((-1/yp1)*(x-x1)+y1,0,1,color="green",aspect_ratio=1);
plot0+plot1+plot2+plot3+plot4
```
