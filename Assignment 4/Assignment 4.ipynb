## Exercici (3) -Reaccions-

```python
var("a, b, c, A0, B0, C0, k, t")
A=function("A")(t)
B=function("B")(t)
C=function("C")(t)
D=function("D")(t)
A0=50
B0=30
C0=20
a=4
b=2
c=2       
```

```python
a+b+c
```

```python
eq1= diff(A,t)==-a*k*A*B*C
eq2= diff(B,t)==-b*k*A*B*C
eq3= diff(C,t)==-c*k*A*B*C
eq4= diff(D,t)==(a+b+c)*k*A*B*C
```

```python
var("t")
```

Trobem les expressions de com varien A(t), B(t), C(t) en funció de D(t)

```python
eq1A=desolve(eq4/eq1,D,ivar=t,ics=[0,0],contrib_ode=True)
sol=solve(D==eq1A, A); sol
```

```python
solA= solve(D==eq1A, A)[0].rhs(); solA
```

Ara fem el mateix per B i C

```python
eq2B=desolve(eq4/eq2,D,ivar=t,ics=[0,0],contrib_ode=True)
solB= solve(D==eq2B, B)[0].rhs(); solB
```

```python
eq3C=desolve(eq4/eq3,D,ivar=t,ics=[0,0],contrib_ode=True)
solC= solve(D==eq3C, C)[0].rhs(); solC
```

Ara escribim les funcions substituint A(0), B(0), C(0) per A0, B0, C0 respectivament

```python
eq11= A0 - 1/2*D(t)
eq21= B0 - 1/4*D(t)
eq31= C0 - 1/4*D(t)
```

```python
eq41=eq4.subs(A(t)==eq11, B(t)==eq21, C(t)==eq31)
eq41
```

Resolem la D

```python
solD=desolve(eq41,D,ivar=t,ics=[0,0])
solD.show()
```

```python
solDD=solD.simplify_log()*200*k
show(solDD)
```

```python
solsD=solve(solDD,D)[0].rhs()
show(solsD)
```
