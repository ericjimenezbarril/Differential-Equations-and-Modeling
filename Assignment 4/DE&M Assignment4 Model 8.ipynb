```python
# Model 8
# Determineu la família de corbes ortogonals al feix {(x,y) dins  R^2: y=e^{c x +5}, c és dins R }.
```

```python
# Per cada c, tenim una corba y=e^{c x +5}
```

```python
var("c x");fun=e^(c*x +5)
```

```python
fun(x=1,c=-2)
```

```python
plot1=plot(fun(c=1),-2,1);plot1
```

```python
plot2=plot(fun(c=2),-2,1);
plot0=plot(fun(c=0),-2,1);
plot3=plot(fun(c=-1),-2,1);plot0+plot1+plot2+plot3
```

```python
# Busquem una família de corbes que en tot punt siguin ortogonals a una de la família anterior.
```

```python
# Aïllem la constant
```

```python
var("y");fun_imp=solve(y==fun,c);fun_imp
```

```python
# Aquesta és la funció implícita c=F(x,y)
#
# Calculem la y'= -Fx/Fy

```

```python
Fx=diff(fun_imp[0].rhs(),x);Fx
```

```python
Fy=diff(fun_imp[0].rhs(),y);Fy
```

```python
# Per tant, si la tangent és m=-Fx/Fy, l'ortogonal tindrà tangent Fy/Fx. Busquem doncs una família de funcions 
# amb aquesta derivada
```

```python
eq1=Fy/Fx; eq1
```

```python
# La resolem
```
