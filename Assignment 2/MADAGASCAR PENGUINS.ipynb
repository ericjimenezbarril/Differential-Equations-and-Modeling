# Primer model

### Estudi de la temperatura de l'interior de l'habitatge d'un dia de l'any D

#### Temperatura exterior Manhattan d'un dia

```python
D=54
```

```python
Te(t)=(Tmax[D]+Tmin[D])/2 + (Tmax[D]-Tmin[D])/2 * sin(-2*pi*t/24)
```

#### Interval 1: [0,t1]

Fase en que els pingüins no estan, no s'ha de refrigerar l'habitatge: $q(t)=0$. Condició inicial: $T_1(0)=T_e(0,D)$

```python
T1=function("T1")(t)
solT1=desolve(diff(T1(t),t)+k*(T1(t)-Te(t)),[T1,t],[0,Te(0)])
```

#### Interval 2: [t1,10]

Fase en que els pingüins estan a punt d'arribar, s'ha de començar a refrigerar l'habitatge per tal que a les 10h estigui a la temperatura adequada (1 grau): $q(t)=M$ ($M:=$ màxima potència del refrigerador. Condició inicial: $T_2(10)=1$

```python
T2=function("T2")(t)
solT2=desolve(diff(T2(t),t)==-k*(T2(t)-Te(t))-M,[T2,t],[10,1])
```

#### Interval 3: [10,11]

Fase en que els pingüins estan a dins l'habitatge, volem aconseguir una temperatura constantment 1: $q(t)=k(1-T_e(t,D))$. Condició inicial: $T_3(10)=1$

```python
solT3(t)=1
```

#### Interval 4: [11,t2]

Fase en que els pingüins marxen de l'habitatge, no s'ha de refrigerar: $q(t)=0$. Condició inicial: $T_4(11)=1$

```python
T4=function("T4")(t)
solT4=desolve(diff(T4(t),t)+k*(T4(t)-Te(t)),[T4,t],[11,1])
```

#### Interval 5: [t2,13]

Fase en que els pingüins estan a punt d'arribar, s'ha de començar a refrigerar l'habitatge per tal que a les 13h estigui a la temperatura adequada (1 grau): $q(t)=M$. Condició inicial: $T_5(13)=1$

```python
T5=function("T5")(t)
solT5=desolve(diff(T5(t),t)==-k*(T5(t)-Te(t))-M,[T5,t],[13,1])
```

#### Interval 6: [13,17]

Fase en que els pingüins estan a dins l'habitatge, volem aconseguir una temperatura constantment 1: $q(t)=k(1-T_e(t,D))$. Condició inicial: $T_6(13)=1$

```python
solT6(t)=1
```

#### Interval 7: [17,24]

Fase en que els pingüins marxen de l'habitatge, no s'ha de refrigerar: $q(t)=0$. Condició inicial: $T_7(17)=1$

```python
T7=function("T7")(t)
solT7=desolve(diff(T7(t),t)+k*(T7(t)-Te(t)),[T7,t],[17,1])
```

#### Definició de t1 i t2

Són els temps en que $T_1=T_2$ i $T_4=T_5$:

```python
t1=N(solve([solT1==solT2],t)[0].rhs())
t2=N(solve([solT4==solT5],t)[0].rhs())
```

#### Funció T(t) := temperatura interior de l'habitatge de pinguins durant el dia D

```python
def T(t):
    if 0<=t<t1:
        return solT1(t)
    if t1<=t<10:
        return solT2(t)
    if 10<=t<11:
        return solT3(t)
    if 11<=t<t2:
        return solT4(t)
    if t2<=t<13:
        return solT5(t)
    if 13<=t<17:
        return solT6(t)
    if 17<=t<=24:
        return solT7(t)
```

```python
plot(T,(0,24))
```

### Energia consumida en un dia D de l'any

El cost energètic total d'un dia de l'any serà proporcional a l'energia consumida de tot el dia

#### Energia consumida als intervals 1, 4 i 7

És 0, tenim la font de calor apagada.

#### Energia consumida als intervals 2 i 5

La potència és màxima ($M$)

```python
E1=-M*(10-t1)-M*(13-t2)
```

#### Energia consumida als intervals 3 i 6

La potència és $q(t)=k(1-T_e(t,D))$

```python
E2 = integral(k*(1-Te(t)),t,10,11) + integral(k*(1-Te(t)),t,13,17)
```

#### Energia total d'un dia

```python
E=abs(E1)+abs(E2)
N(E)
```

### Energia total anual

```python
Etot=0
for D in [0..364]:
    Te(t)=(Tmax[D]+Tmin[D])/2 +(Tmax[D]-Tmin[D])/2 * sin(-2*pi*t/24)
    T1=function("T1")(t)
    solT1=desolve(diff(T1(t),t)+k*(T1(t)-Te(t)),[T1,t],[0,Te(0)])
    T2=function("T2")(t)
    solT2=desolve(diff(T2(t),t)==-k*(T2(t)-Te(t))-M,[T2,t],[10,1])
    T4=function("T4")(t)
    solT4=desolve(diff(T4(t),t)+k*(T4(t)-Te(t)),[T4,t],[11,1])
    T5=function("T5")(t)
    solT5=desolve(diff(T5(t),t)==-k*(T5(t)-Te(t))-M,[T5,t],[13,1])
    t1=N(solve([solT1==solT2],t)[0].rhs())
    t2=N(solve([solT4==solT5],t)[0].rhs())
    Etot=Etot+abs(-M*(10-t1)-M*(13-t2))+abs(integral(k*(1-Te(t)),t,10,11)) + abs(integral(k*(1-Te(t)),t,13,17))
show(N(Etot))
```

Observem diverses millores en aquest primer model més simple:


 1. És irreal tenir una temperatura constantment 1º en les hores en que estan els pingüins; poden sobreviure en un interval de temperatures a prop de 1º. Això de mantenir estrictament constant la temperatura només es té en laboratoris i quirofans.
 
 
 2. És molt difícil controlar la quantitat de calor que subministres a l'habitatge depenent la temperatura exterior. S'hauria de tenir un termòmemtre a l'exterior que enviés les dades al moment per poder decidir la quantitat de calor.
 
 
 3. A part, és poc econòmic.
 
 
 4. En els intervals on volem refredar l'habitatge perquè en una certa hora estigui a la temperatura que volem, estem utilitzant la màxima potència. Això també és una despesa d'energia molt alta que es podria minimitzar.
